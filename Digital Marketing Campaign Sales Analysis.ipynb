### (v)Creating Derived Columns
In this step,new columns such as **Income Group**, **Age Group** and **Customer Value Score** are created to organize the data into clearer categories for analysis.

```python
print(df.columns.tolist())
```

```python
#Creating Column named Age Group
df['Age Group'] = pd.cut(
    df['Age'],
    bins=[18,25,35,45,55,65,75],
    labels=['18-25','26-35','36-45','46-55','56-65','66-75'],
    include_lowest=True
)
```

***Info:-*** A new column named **Age Group** was created to categorize customers into different age ranges.

```python
# Define the edges of your bins (include a bit below the min and above the max)
income_bins = [20000, 50000, 100000, 150000]
income_labels = ['Low', 'Medium', 'High']

df['Income Group'] = pd.cut(
    df['Income'],
    bins=income_bins,
    labels=income_labels
)
```

***Info:-*** A new column named **Income Group** was created using `pd.cut()` with manually defined bins (₹20,000–₹50,000 → Low, ₹50,000–₹1,00,000 → Medium, ₹1,00,000–₹1,50,000 → High) to segment customers into three income categories based on predefined income thresholds.

```python
#Creating Column named Customer Value Score
df['Customer Value Score'] = (
    df['Previous Purchases'] +
    (df['Loyalty Points'] / 100)
)
```

***Info:-*** A new column named **Customer Value Score** was created using Previous Purchases and Loyalty Points to estimate the overall value of a customer. Higher scores indicate customers with stronger purchasing history and greater loyalty, helping identify the most valuable customer segments.

#### (iii) Bivariate Analysis
It examines the relationship between two variables to identify patterns,associations,and differences. It helps understand how one feature may influence or vary with another.

```python
# Campaign Channel vs Conversion
def Conversion_Stats(column):

    result = df.groupby(column).agg(
        Total_Conversions=('Conversion Status','sum'),
        Conversion_Rate=('Conversion Status',
                         lambda x: round(x.mean()*100,2))
    )

    return result
```

***Info:-*** Groups the data by a chosen column by using groupby function and calculates the conversion count and conversion percentage for each group.

```python
Conversion_Stats('Campaign Channel')
```

***Info:-*** Referral campaigns achieved the highest conversion rate (88.31%), closely followed by PPC campaigns (88.28%). Social Media campaigns recorded the lowest conversion rate (86.83%), while all channels demonstrated consistently high conversion performance above 86%.

```python
Conversion_Stats('Age Group')
```

***Info:-*** Customers aged 36–45 recorded the highest conversion rate (90.14%), indicating stronger responsiveness to marketing campaigns. The lowest conversion rate was observed in the 66–75 age group (86.06%), while other age groups showed relatively similar performance.

```python
Conversion_Stats('Income Group')
```

***Info:-*** Conversion rates were fairly consistent across all income groups, ranging from 87.29% to 88.15%. High-income customers showed the highest conversion rate, suggesting a slightly greater likelihood of conversion compared to other income segments.

```python
Conversion_Stats('Gender')
```

***Info:-*** Male and female customers exhibited nearly identical conversion rates, with males at 87.69% and females at 87.62%. This indicates that gender has minimal influence on customer conversion behavior.

```python
Conversion_Stats('Campaign Type') 
```

***Info:-*** Conversion campaigns achieved the highest conversion rate (93.36%), significantly outperforming Awareness, Consideration, and Retention campaigns, which recorded conversion rates around 85–86%. This suggests that campaigns specifically designed for conversion are the most effective in generating customer actions.

***Info:-*** This analysis is considered bivariate because it explores the relationship between a categorical variable and Conversion Status. The aggregated metrics help compare conversion performance across different customer segments and marketing categories.

### (iv) Multivariate Analysis
It examines how multiple variables interact simultaneously to uncover deeper patterns,combined effects, and factors influencing the target outcome.

```python
# Identifying how different age groups respond to various campaign channels

df.groupby(
    ['Campaign Channel','Age Group'],
    observed=True
).agg(
    Average_Conversion_Rate=('Conversion Rate','mean')
).round(2)
```

***Info:-*** This analysis shows that average conversion rates remain relatively consistent across campaign channels and age groups, ranging between 10% and 11%. Slightly higher conversion rates were observed for Social Media and Email campaigns in certain age segments, indicating a marginally better performance among those customer groups.

```python
# Examining conversion performance across Income Group and Gender segments

df.groupby(
    ['Income Group','Gender'],
    observed=True
).agg(
    Average_Conversion_Rate=('Conversion Rate','mean')
).round(2)
```

***Info:-*** The average conversion rates are very similar across income and gender segments, varying between 10% and 11%. High-income customers show slightly higher conversion rates regardless of gender, suggesting a minor positive relationship between income level and conversion performance.

```python
# Evaluating the combined impact of Campaign Channel and Campaign Type on Website Visits

df.groupby(
    ['Campaign Channel','Campaign Type'],
    observed=True
).agg(
    Average_Website_Visits=('Website Visits','mean')
).round(2)
```

***Info:-*** Website engagement varies across campaign channel and campaign type combinations. SEO-Consideration campaigns generated the highest average website visits, while PPC-Conversion campaigns recorded the lowest. Overall, Consideration campaigns tend to attract greater customer engagement across multiple channels.

```python
# Analyzing how Age Group and Income Group together influence Customer Value Score

df.groupby(
    ['Age Group','Income Group'],
    observed=True
).agg(
    Average_Customer_Value_Score=('Customer Value Score','mean')
).round(2)
```

***Info:-*** Customer Value Scores remain relatively stable across all age and income segments, generally ranging between 28 and 30. Slightly higher scores are observed among older customers in the low-income group, indicating marginal differences in customer value across demographic segments.

```python
# Analyzing how Campaign Channels differ in terms of Ad Spend, Website Visits, and Conversion Rate

df.groupby(
    'Campaign Channel',
    observed=True
).agg(
    Average_Ad_Spend=('Ad Spend', 'mean'),
    Average_Website_Visits=('Website Visits', 'mean'),
    Average_Conversion_Rate=('Conversion Rate', 'mean')
).round(2)
```

***Info:-*** The analysis indicates that marketing performance is relatively consistent across campaign channels. While Social Media generated the highest average website visits and conversion rate, differences between channels are small, suggesting comparable effectiveness across marketing channels.
